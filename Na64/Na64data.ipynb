```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
def plot_curve(data: pd.DataFrame, sample_name=None):
    fig = plt.figure(figsize=(8, 8))
    if sample_name:
        fig.suptitle(sample_name, fontsize=16)

    gs = gridspec.GridSpec(2, 2, height_ratios=[2, 1])  # 2 rows, 2 columns
    ax1 = fig.add_subplot(gs[0, :])
    ax1.plot(data.ma_prime, data.epsilon, color='blue', linewidth=2, label='Closed Line')
    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.set_title("Curve")

    ax2 = fig.add_subplot(gs[1, 0])
    ax2.plot(data.index, data.ma_prime, color='red', lw=2)
    ax2.set_title("ma_prime vs index")

    ax3 = fig.add_subplot(gs[1, 1])
    ax3.plot(data.index, data.epsilon, color='red', lw=2)
    ax3.set_title("epsilon vs index")
```

```python
# Read dataframe from csv
na64_eminus_data = pd.read_csv('NA64e-_2022.csv', header=None, names=['ma_prime', 'epsilon'])
na64_eplus_data = pd.read_csv('NA64e+_2022.csv', header=None, names=['ma_prime', 'epsilon'])[:-1]
na64_mu2022_data = pd.read_csv('NA64mu_2022.csv', header=None, names=['ma_prime', 'epsilon'])
na64_mu3e11_data = pd.read_csv('NA64mu_3e11.csv', header=None, names=['ma_prime', 'epsilon'])
na64_mu1e14_data = pd.read_csv('NA64mu_1e14.csv', header=None, names=['ma_prime', 'epsilon'])
na64_mu2e13_data = pd.read_csv('NA64mu_2e13.csv', header=None, names=['ma_prime', 'epsilon'])
babar_data = pd.read_csv('BaBar.csv', header=None, names=['ma_prime', 'epsilon'])
```

```python

plot_curve(na64_eminus_data, "NA62 e-")
plot_curve(na64_eplus_data, "Na62 e+")
plot_curve(na64_mu2022_data, "Na62 mu 2022")
plot_curve(na64_mu3e11_data, "Na62 mu 3e11")
plot_curve(na64_mu1e14_data, "Na62 mu 1e14")
plot_curve(na64_mu2e13_data, "Na62 mu 2e13")
plot_curve(babar_data, "Babar")
```
